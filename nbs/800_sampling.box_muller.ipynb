Box-Muller Algorithm
1. Generate $U1∼uniform(0,1)$ and $U2∼uniform(0,1)$ where $U1⊥U2$
2. Set $R=−2log(U1)$ and $\theta=2πU2$
3. Set $X=Rcos(\theta)$ and $Y=Rsin(\theta)$

```python
#| export

def sample(size=1000):
    U1 = np.random.uniform(size = size)
    U2 = np.random.uniform(size = size)
    R = np.sqrt(-2 * np.log(U1))
    theta = 2 * np.pi * U2
    X = R * np.cos(theta)
    # Y = R * np.sin(theta)
    return X
```

```python
#| include: false
import numpy as np
import pandas as pd
import scipy.stats as stats
import seaborn as sns
import matplotlib.pyplot as plt

np.random.seed(521)

X = sample()

def p(x, mu=0, scale=1):
    return stats.norm.pdf(x, loc=mu, scale=scale)

sns.displot(X, kind='kde', color='r')
plt.scatter(X, p(X))
plt.show()
```
